```python

#text = '예술강사'
df = pd.read_csv('instaTagArea_v0.1.csv', na_values='-').drop_duplicates(subset='URL',keep='first')
df.fillna(0)
tagStr = df[['URL','HASH','UPDATE_DATE','SEARCH_TEXT','CONTENT','AREA']].replace(np.nan,'')
artarr = []

for i in tagStr.index:
    artarr.append([tagStr.loc[i]['AREA'],tagStr.loc[i]['HASH'].split(',')])
result = analDivision(artarr)   #전국 17개 도,시 구분
resultList = tagCnt(result)  #구분된 17개 도,시별 해시태그 카운트
print(resultList)
data = pd.DataFrame(sorted(resultList,key=itemgetter(1), reverse=True), columns=["HASH","COUNT"])  #DataFrame을 이용하여 엑셀데이터의 컬럼명을 추가(데이터의 컬럼의 수와 컬럼명의 수가 일치해야함)
excelWriter = pd.ExcelWriter('HASH_.xlsx', engine='xlsxwriter', options={'strings_to_urls': False})     #엑셀데이터 export시 제목 설정
data.to_excel(excelWriter)
excelWriter.close()
```
